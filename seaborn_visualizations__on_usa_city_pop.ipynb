# **Visualizations**

```python
df=pd.DataFrame(city.head(10))
```

Line plot

```python
plt.figure(figsize=(30,4))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x=df['name'],y=df['growth'],data=df)
plt.show()
```

Bar plot

```python
plt.figure(figsize=(15,6))
sns.barplot(data=df,x=df['name'],y=df['growth'])
```

```python
from numpy import median
sns.set(rc={"figure.figsize":(20, 4)})
ax=sns.barplot(x="name", y="pop2010", data=df,  palette="Blues_d");
```

scatter plot

```python
plt.figure(figsize=(20,6))
sns.scatterplot(data=df,x=df['name'],y=df['growth']);
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(data=df,x=df['name'],y=df['growth'],sizes=(15, 200),size="pop2022");
```

```python
df.columns
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(data=df,x=df['name'],y=df['growth'],sizes=(15, 200),size="pop2022",marker="+",color='black',s=(200));
```

```python
df.columns
```

Distribution plot

```python
sns.displot(data=df,x=df['growth'],y=df['density'])  
```

KDE plot

```python
sns.displot(data=df,x=df['growth'],y=df['density'],kind='kde')  
```

```python
sns.displot(data=df,x=df['growth'],y=df['density'],kind='kde',log_scale=(True, False)) 
```

Joint plot

```python
sns.jointplot(data=df,x=df['growth'],y=df['density'])
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(data=df,x=df['pop2022'],y=df['pop2010'],sizes=(15, 200),size="pop2022");
sns.rugplot(data=df,x=df['pop2022'],y=df['pop2010']);
```

```python
sns.displot(
    df, x="growth", y="density",
    kind="kde", rug=True
)
```

Pairplot

```python
sns.pairplot(df)
```

```python
g = sns.PairGrid(df)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```

Histogram

```python
sns.histplot(data=df,x=df['density'],kde=True);
```

```python
sns.histplot(data=df);
```
